### Problem 1

```python

# Problem Data
N = np.array([0, 4, 2, 2, 3, 0, 4, 5, 6, 6, 4, 1,
              4, 4, 0, 1, 3, 4, 2, 0, 3, 2, 0, 1])
rho = np.array([0.1, 1, 10, 100])

# Optimization Variable
lamda = cvx.Variable(24)

# Solve Optimization Problem for 4 different rhos
for i in rho:
    objective = cvx.Minimize(cvx.sum(lamda)
                             - cvx.sum(cvx.multiply(N, cvx.log(lamda)))
                             + i * cvx.sum_squares(cvx.diff(lamda))
                             + i * (lamda[0]-lamda[23])**2)
    problem = cvx.Problem(objective)
    problem.solve()

    # Plot the optimum points
    plt.plot(np.arange(24), lamda.value, label="%.1f" % i)
# Plot the first Datas
plt.plot(np.arange(24), N)

# Show PLots
plt.legend()
```

```python
constraints = [cvx.inv_pos(x) + cvx.inv_pos(y) <= 1]
constraints += [x >= 0, y >= 0]
problem = cvx.Problem(objective, constraints)
print(problem.is_dcp())
```

```python
constraints = [-x + cvx.inv_pos(y) <= 0]
constraints += [x >= 0, y >= 0]
problem = cvx.Problem(objective, constraints)
print(problem.is_dcp())
```
